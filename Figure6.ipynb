```python
import numpy as np    
from flyeye.data.experiments import Experiment
from figures.dynamics import DualExpression
from figures.comparison import PerturbationComparison
%matplotlib inline
```

# Load SevRasV12 experiment data

```python
ras_control = Experiment('../data/ras_mutant/control', normalization='red')
ras_perturbation = Experiment('../data/ras_mutant/perturbation', normalization='red')

# manually align control condition (falls ~10h early by default)
ras_control.apply_lag(lag=10)
```

# Figure 6 A-C

```python
# define number of bootstrap samples (10k used for manuscript)
interval_kw = dict(nbootstraps=10) 

# Fig 6A (Pnt levels)
fig6A = DualExpression(ras_control, ras_perturbation)
ax_kw = dict(xlim=(-20, 50), xticks=np.arange(-20, 55, 10), 
             ylim=(.2, 1.2), yticks=np.arange(.2, 1.3, .3))
fig6A.render('green', interval=True, ax_kw=ax_kw, interval_kw=interval_kw)

# Fig 6B (Yan levels)
fig6B = DualExpression(ras_control, ras_perturbation)
ax_kw = dict(xlim=(-20, 50), xticks=np.arange(-20, 55, 10), 
             ylim=(.2, 1.2), yticks=np.arange(.2, 1.3, .3))
fig6B.render('blue', interval=True, ax_kw=ax_kw, interval_kw=interval_kw)

# Fig 6C (Ratio)
fig6C = DualExpression(ras_control, ras_perturbation)
ax_kw = dict(xlim=(-20, 50), xticks=np.arange(-20, 55, 10), 
             ylim=(-1, 0.7), yticks=np.arange(-1, .6, .5))
fig6C.render('ratio', interval=True, ax_kw=ax_kw, interval_kw=interval_kw)
```

# Figure 6 D,E

```python
# Fig 6D (unidentifiable supernumerary neurons)
fig6D = PerturbationComparison.from_experiments(ras_control, 
                                                ras_perturbation, 
                                                reference=['er'])
fig6D.render()

# Fig 6E (supernumerary R7 neurons)
fig6E = PerturbationComparison.from_experiments(ras_control, 
                                                ras_perturbation, 
                                                reference=['er7'])
fig6E.render()
```
